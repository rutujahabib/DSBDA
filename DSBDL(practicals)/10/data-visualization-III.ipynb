```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris') 
```

```python
# 1. List down features and their types
print("1. Features and their types")
print(iris.dtypes)
print("\nFeature Types:")
for column in iris.columns:
    dtype = iris[column].dtype
    if dtype == 'object':
        print(f"{column}: Nominal")
    else:
        print(f"{column}: Numeric")
```

```python
# 2. Histogram for each numeric feature
numeric_features = iris.select_dtypes(include='number').columns
print("\n2. Plotting histograms...")
iris[numeric_features].hist(figsize=(10, 8), color='lightblue', edgecolor='black')
plt.suptitle('Histograms of Iris Features')
plt.tight_layout()
plt.show()
```

```python
print("3. Plotting boxplots...")
plt.figure(figsize=(10, 8))
for i, feature in enumerate(numeric_features, 1):
 plt.subplot(2, 2, i)
 sns.boxplot(y=iris[feature], color='lightgreen')
 plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()
```

```python
# 4. Compare distributions and identify outliers (inference)
print("4. Inference:")
for feature in numeric_features:
 q1 = iris[feature].quantile(0.25)
 q3 = iris[feature].quantile(0.75)
 iqr = q3 - q1
 lower = q1 - 1.5 * iqr
 upper = q3 + 1.5 * iqr
 outliers = iris[(iris[feature] < lower) | (iris[feature] > upper)]
 print(f"- {feature}: {len(outliers)} outlier(s) detected.")
```
